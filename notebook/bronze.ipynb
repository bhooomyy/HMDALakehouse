```python
import urllib.request
```

```python
years = range(2013, 2018)   # 2013 through 2017 -- last 5 years

for year in years:
    files = {
        "panel": f"https://files.consumerfinance.gov/hmda-historic-institution-data/hmda_{year}_panel.zip",
        "lar":   f"https://files.consumerfinance.gov/hmda-historic-loan-data/hmda_{year}_nationwide_first-lien-owner-occupied-1-4-family-records_labels.zip",
    }
    for folder, url in files.items():
        fname = f"hmda_{year}_{folder}.zip"
        urllib.request.urlretrieve(url, f"/Volumes/hmdaproj/default/bronze/{fname}")
        dbutils.fs.cp(
            f"/Volumes/hmdaproj/default/bronze/{fname}",
            f"abfss://bronze@hmdaprojstorage.dfs.core.windows.net/{folder}/{fname}"
        )
        print(f"{year} {folder} done")
```
